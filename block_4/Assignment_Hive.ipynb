## Ex 2
Get ``family_name`` and ``salary`` of employees sorted by salary. Return first 100 entries. Hint: you need to join both tables...

```python
cursor.execute("SELECT e.field_4, s.field_2 FROM hue__tmp_employees e JOIN hue__tmp_salaries s ON e.field_1 = s.field_1 ORDER BY s.field_2 LIMIT 100")
data=cursor.fetchall()
print(data)
```

## Ex 3
Get the average salary by gender. Hint: use Group by

```python
cursor.execute("SELECT e.field_5, AVG(s.field_2) FROM hue__tmp_employees e JOIN hue__tmp_salaries s ON e.field_1 = s.field_1 GROUP BY e.field_5")
print(cursor.fetchall())
```
